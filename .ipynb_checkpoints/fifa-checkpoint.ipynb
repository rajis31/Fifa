Below we are going to create a function to plot a spider plot for any player based on any number of numeric characteristics. This will work on missing values. 

```python
def player_spider_plot(data, metrics,player_id):

    nmetrics=len(metrics)
    angles=[n / float(nmetrics) * 2 * np.pi for n in range(nmetrics)]
    angles.append(0)
    values=data[data["ID"]==player_id][metrics].values.tolist()[0]
    values.append(values[0])
    name=data[data["ID"]==player_id]["Name"][0]

    ax=plt.subplot(111,polar=True)
    plt.xticks(angles[:-1],metrics)
    ax.plot(angles,values)
    ax.fill(angles, values, 'teal', alpha=0.1)
    ax.set_title(name,fontsize=16,fontweight="bold")
    plt.show()
```

We will start with using linear regression to predict wage below. 

```python
### Separate DF into independent (exogenous) and dependent (endogenous) variable
## Drop some columns in X
## Drop club as well. Will make the data wide. 
## Drop all the columns between LS and RB since these are based on ratings we don't need
X=fifa.drop(columns=["Name","Photo","Club Logo","Jersey Number","Contract Valid Until","Wage Unit","Wage","Flag","Nationality","Flag","Club"])
X.drop(columns=X.columns.to_series()["LS":"RB"],inplace=True)
y=fifa[["Wage"]] 
```

```python
### 1 hot encode some categorical columns 
dummy1=pd.get_dummies(X[["Position Description"]])
dummy2=pd.get_dummies(X[["Preferred Foot"]])
dummy3=pd.get_dummies(X[["Work Rate"]])
dummy4=pd.get_dummies(X[["Real Face"]])
dummy5=pd.get_dummies(X[["Position"]])
X=pd.concat([X,dummy1,dummy2,dummy3,dummy4,dummy5],axis=1)
```

```python
X.drop(columns=["Position Description","Joined","Loaned From","Release Clause","Value Unit","Body Type","Preferred Foot","Work Rate","Real Face","Position"],inplace=True)
```

```python
X.dtypes[X.dtypes==np.object]
```

```python
#X.columns.to_series().to_csv("columns.csv")
```

```python
X[["Height"]] # Deal with this 
X[["Weight"]]=X["Weight"].str.replace("lbs","").fillna(0).astype(np.int64)
```

```python
mean_weight=int(X.Weight.mean())
X["Weight"]=X["Weight"].apply(lambda x: mean_weight if x==0 else x)
X["value"]=X["value"].apply(lambda x: float((x[:-1]) if x[-1] in ("M","K") else float(X)))
```

```python
def hgt_to_inches(x):
    x=str(x)
    if "'" in x:
        return 12*int(x[:x.index("'")])+int(x[x.index("'")+1:])
    else:
        return 0
hgt_to_inches("5'9")
```

```python
X["Height"]=X["Height"].apply(hgt_to_inches)
X=X.dropna()
y=y.merge(X,how="inner",left_index=True, right_index=True)["Wage"]
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression().fit(X,y)
X["predicted"]=lr.predict(X)
X["Wage"]=y
```

In the result frame below, we see the predicted value and the actual wages.  The predicted values are

Next we will use K-Modes clustering to try to group together similar players. The reason why we use K-Modes rather then K-means is that we have a mix of categorical and continous data types and K-Means is based one euclidean distance which does not make sense to use with categorical data. 

```python
from kmodes import kmodes as k

km = k.KModes(n_clusters=10, init='Huang', n_init=5, verbose=1)
clusters = km.fit_predict(X)

# Print the cluster centroids
print(km.cluster_centroids_)
```
