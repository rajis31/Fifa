Below we are going to create a function to plot a spider plot for any player based on any number of numeric characteristics. This will work on missing values. 

```python
def player_spider_plot(data, metrics,player_id):

    nmetrics=len(metrics)
    angles=[n / float(nmetrics) * 2 * np.pi for n in range(nmetrics)]
    angles.append(0)
    values=data[data["ID"]==player_id][metrics].values.tolist()[0]
    values.append(values[0])
    name=data[data["ID"]==player_id]["Name"][0]

    ax=plt.subplot(111,polar=True)
    plt.xticks(angles[:-1],metrics)
    ax.plot(angles,values)
    ax.fill(angles, values, 'teal', alpha=0.1)
    ax.set_title(name,fontsize=16,fontweight="bold")
    plt.show()
```

**Salary**

In this section we are going to see if we can find a pattern with how much teams spend on their rosters. We begin with a plot below we see that the youngest teams are in the European league while some of the oldest teams are in South American region. We will look to see how much their rosters cost accross these teams to see if we can spot a difference. 

```python
mean_age=np.mean(fifa["Age"]) # Calculate the mean age

avg_age_under=fifa[["Club","Age"]].groupby("Club").mean().apply(lambda x: round(x,1)).sort_values(by="Age",ascending=False).head(10) # Top 10 oldest clubs
avg_age_over=fifa[["Club","Age"]].groupby("Club").mean().apply(lambda x: round(x,1)).sort_values(by="Age").head(10) # Top 10 youngest clubs

avg_age_under["color"]=["blue" if x<mean_age else "red" for x in avg_age_under["Age"]] #code in colors for diverging plot
avg_age_over["color"]=["blue" if x<mean_age else "red" for x in avg_age_over["Age"]]

avg_age_under.reset_index(inplace=True)
avg_age_over.reset_index(inplace=True)


avg_combined=pd.concat([avg_age_under,avg_age_over],ignore_index=True) # Stack both databframes
avg_combined["Age"]=(avg_combined["Age"]-mean_age)/np.std(avg_combined["Age"]) # Standardize age
avg_combined=avg_combined.sort_values(by="Age",ascending=True).reset_index().drop(columns="index") # Sort and reset index so diverging plot can start from the oldest at top to the youngest.

# Plot
plt.figure(figsize=(10,8), dpi= 80) # Adjust plot size
plt.hlines(y=avg_combined.index, xmin=0, xmax=avg_combined.Age, color=avg_combined["color"], alpha=0.8, linewidth=2.5) # Plot horizontal lines
plt.gca().set(xlabel="Z Score", ylabel="Club")
plt.yticks(avg_combined.index, avg_combined["Club"], fontsize=12)
plt.title('Diverging Bars of Avg. Age by Oldest & Youngest Clubs', fontdict={'size':16})
plt.text(1.75,-2.5, "Plot 5",ha="center",fontdict={"color":"red","size":10,"weight":"bold"})
plt.xlim((-2,2))
plt.grid(linestyle='--', alpha=0.5)
plt.show()
```

Here we see that the oldest clubs are generally paid higher then the younger clubs with the exception of ***LOSC LILLE*** which could indicate that the younger clubs may have the potential for more growth compared to some of the older clubs based on age as soccer players generally retire at 35. 

```python
over=fifa[["Club","Age","Wage"]].groupby("Club").agg({"Age":np.mean, "Wage":np.sum}).apply(lambda x: round(x,1)).sort_values(by="Age",ascending=False) \
    .rename(columns={"Age":"Avg. Age","Wage":"Total Wage"}).head(10) # Top 10 oldest clubs
under=fifa[["Club","Age","Wage"]].groupby("Club").aggregate({"Age":np.mean, "Wage":np.sum}).apply(lambda x: round(x,1)).sort_values(by="Age") \
.rename(columns={"Age":"Avg. Age","Wage":"Total Wage"}).head(10) # Top 10 youngest clubs

print("Top 10 Oldest Clubs")
print(over)
print("\nTop 10 Youngest Clubs")
print(under)
```

We will start with using linear regression to predict wage below. 

```python
### Separate DF into independent (exogenous) and dependent (endogenous) variable
## Drop some columns in X
## Drop club as well. Will make the data wide. 
## Drop all the columns between LS and RB since these are based on ratings we don't need
X=fifa.drop(columns=["Name","Photo","Club Logo","Jersey Number","Contract Valid Until","Wage Unit","Wage","Flag","Nationality","Flag","Club","ID"])
X.drop(columns=X.columns.to_series()["LS":"RB"],inplace=True)
y=fifa[["Wage"]] 
```

```python
### 1 hot encode some categorical columns 
dummy1=pd.get_dummies(X[["Position Description"]])
dummy2=pd.get_dummies(X[["Preferred Foot"]])
dummy3=pd.get_dummies(X[["Work Rate"]])
dummy4=pd.get_dummies(X[["Real Face"]])
dummy5=pd.get_dummies(X[["Position"]])
X=pd.concat([X,dummy1,dummy2,dummy3,dummy4,dummy5],axis=1)
```

```python
X.drop(columns=["Position Description","Joined","Loaned From","Release Clause","Value Unit","Body Type","Preferred Foot","Work Rate","Real Face","Position"],inplace=True)
```

```python
X.dtypes[X.dtypes==np.object]
```

```python
#X.columns.to_series().to_csv("columns.csv")
```

```python
X[["Height"]] # Deal with this 
X[["Weight"]]=X["Weight"].str.replace("lbs","").fillna(0).astype(np.int64)
```

```python
mean_weight=int(X.Weight.mean())
X["Weight"]=X["Weight"].apply(lambda x: mean_weight if x==0 else x)
X["value"]=X["value"].apply(lambda x: float((x[:-1]) if x[-1] in ("M","K") else float(X)))
```

```python
def hgt_to_inches(x):
    x=str(x)
    if "'" in x:
        return 12*int(x[:x.index("'")])+int(x[x.index("'")+1:])
    else:
        return 0
hgt_to_inches("5'9")
```

```python
X["Height"]=X["Height"].apply(hgt_to_inches)
X=X.dropna()
y=y.merge(X,how="inner",left_index=True, right_index=True)["Wage"]
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression().fit(X,y)
X["predicted"]=lr.predict(X)
X["Wage"]=y
```

In the result frame below, we see the predicted value and the actual wages.  The predicted values are

Next we will use K-Modes clustering to try to group together similar players. The reason why we use K-Modes rather then K-means is that we have a mix of categorical and continous data types and K-Means is based one euclidean distance which does not make sense to use with categorical data. 

```python
from kmodes import kmodes as k

km = k.KModes(n_clusters=10, init='Huang', n_init=5, verbose=1)
clusters = km.fit_predict(X)

# Print the cluster centroids
print(km.cluster_centroids_)
```
